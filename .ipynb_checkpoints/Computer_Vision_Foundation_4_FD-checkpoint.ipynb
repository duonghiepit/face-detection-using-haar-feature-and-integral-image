```python
def close_window_os():
    cv2.destroyAllWindows()
    for i in range(1, 5): #trick for closing the display window on MacOS
        cv2.waitKey(1)
```

```python
#Problem 1
import os
import cv2
import pickle
import numpy as np

#Read image from the file
image = cv2.imread("vinh.jpg", 1)

facecascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

face_coordinates = facecascade.detectMultiScale(gray, 1.3, 4)

for (a, b, w, h) in face_coordinates:
    cv2.rectangle(image, (a, b), (a+w, b+h), (255, 0, 0), 2)
    
#Question 1:
print("Face detection's coordinate: ", face_coordinates)
cv2.imshow('frames', image)
cv2.waitKey(0)
close_window_os()
```

```python
import os
import cv2
import pickle
import numpy as np

face_data = []
i = 0

camera = cv2.VideoCapture(0)

facecascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
name = input('Enter your name --> ')
ret = True

# Face recognition using KNN
while(ret):
    ret, frame = camera.read()
    if ret == True:
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        
        face_coordinates = facecascade.detectMultiScale(gray, 1.3, 4)
        
        for (a, b, w, h) in face_coordinates:
            faces = frame[b:b+h, a:a+w, :]
            resized_faces = cv2.resize(faces, (50, 50))
            
            if i % 10 == 0 and len(face_data)<10:
                face_data.append(resized_faces)
                
            cv2.retangle(frame, (a, b), (a+w, b+h), (255, 0, 0), 2)
        i += 1
        
        cv2.imshow('frames', frame)
        
        if cv2.waitKey(1) == 27:
            break
            
    else:
        print('error')
        break
        
cv2.destroyAllWindows()
camera.release()
close_window_os()

face_data = np.asarray(face_data)
face_data = face_data.reshape(10, -1)

if 'names.pkl' not in os.listdir('C:/Users/Admin/Face Detection using Haar feature and Integral image/dataset'):
    names = [name]*10
    with open('C:/Users/Admin/Face Detection using Haar feature and Integral image/dataset/names.pkl', 'wb') as file:
        pickle.dump(names, file)
else:
    with open('C:/Users/Admin/Face Detection using Haar feature and Integral image/dataset/names.pkl', 'rb') as file:
        names = pickle.load(file)
        
    names = names + [name]*10
    with open('C:/Users/Admin/Face Detection using Haar feature and Integral image/dataset/names.pkl', 'wb') as file:
        pickle.dump(names, file)
        
if 'faces.pkl' not in os.listdir('C:/Users/Admin/Face Detection using Haar feature and Integral image/dataset'):
    with open('C:/Users/Admin/Face Detection using Haar feature and Integral image/dataset/faces.pkl', 'wb') as w:
        pickle.dump(face_data, w)
else:
    with open('C:/Users/Admin/Face Detection using Haar feature and Integral image/dataset/faces.pkl', 'rb') as w:
        faces = pickle.load(w)

    faces = np.append(faces, face_data, axis=0)
    with open('C:/Users/Admin/Face Detection using Haar feature and Integral image/dataset/faces.pkl', 'wb') as w:
        pickle.dump(faces, w)
```
